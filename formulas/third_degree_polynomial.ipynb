```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 3, 201)
print(np.poly([1, 1, 2]))
plt.plot(x, x**3 - 4*x**2 + 5*x - 2)
plt.plot(x, np.polyval([1., -4., 5., -2], x))
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2, 2)
plt.grid(color='#eeefff')
print(np.roots([1., -4., 5., -2.]))
print(list(map(lambda x: round(x, 2), np.roots([1, -4, 5, -2]))))
```
