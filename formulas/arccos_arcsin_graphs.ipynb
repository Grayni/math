```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 201)
plt.plot(x, np.arccos(x), color='blue')
plt.plot(x, np.arcsin(x), color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-1, 1)
plt.ylim(-np.pi, np.pi)
plt.plot([-1, 1], [-np.pi/2, -np.pi/2], color='#555fff', linewidth=0.75, linestyle='--')
plt.plot([-1, 1], [0, 0], color='#444fff', linewidth=0.75, linestyle='--')
plt.plot([-1, 1], [np.pi/2, np.pi/2], color='#555fff', linewidth=0.75, linestyle='--')
plt.grid(color='#eeefff')
```
