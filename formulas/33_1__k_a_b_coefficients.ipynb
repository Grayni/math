```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10, 7))
x = np.linspace(-3 * np.pi, 3 * np.pi, 201)
plt.plot(x, np.cos(x), color='red')  # standart
plt.plot(x, 2 * np.cos(x), color='pink')  # stretching in height ↕
plt.plot(x, np.cos(x - 0.4), color='#888fff') # negative → ; positive ←
plt.plot(x, np.cos(x) + 1, color='#888')  # negative ↓ ; positive ↑
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color='#eeefff')
```
