$\mathbf{Решить \ уравнение}$

$$sin(а*x)=0$$

$$0.01<a<0.02, \ 100<х<500$$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
a_range = np.linspace(0.01 + 0.0002, 0.02 - 0.0002, 50)
x_range = np.linspace(100 + 0.01, 500 - 0.01, 400)
min_ax = np.amin(a_range) * np.amin(x_range)  # 1
max_ax = np.amax(a_range) * np.amax(x_range)  # 10

print(min_ax, max_ax)  # create range pi

# sin(value) = 0  # 0, pi, 2*pi, 3 * pi, 4 * pi ...

min_pi = min_ax / np.pi  # 0.3183 value => pi * 0.3183 = 1
max_pi = max_ax / np.pi  # 3.1831 value => pi * 3.1831 = 10
#  (0.3183*pi ... 3.1831*pi)

# sin(value) = 0  # value =: pi, 2*pi, 3 * pi

arr_pi_values = []
sum_pi = 0

pi_count = 0

while True:
    sum_pi += np.pi
    if min_ax <= sum_pi <= max_ax:
        pi_count += 1
        arr_pi_values.append(sum_pi)
    else:
        break

print('List pi * n:', arr_pi_values, end='\n\n')  # [3.141592653589793, 6.283185307179586, 9.42477796076938] 3 values


def x(a, pi_n):
    # pi * n / a
    return np.pi * pi_n/a



print(f'Example: a=0.0102 * x=307.9992797637052 = pi*n = {0.0102 * 307.9992797637052} => sin(pi*n) = 0', end='\n\n')
a_res = []
value_res = []
result = []
for num in range(1, pi_count + 1):
    result.append([])
    for a in a_range:
        value = x(a, num)
        if 100 < value < 500:
            a_res.append(a)
            value_res.append(value)
            result[num-1].append([a, value])
    plt.plot(a_res, value_res)
    a_res = []
    value_res = []

print(result)





# print(0.02*471.23889803846896)
# print(round(np.sin(9.42477796076938), 10))

# plt.plot(a_res, value_res)
```
