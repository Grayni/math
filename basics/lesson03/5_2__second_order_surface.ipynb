### Эллиптический параболоид

$$z = \frac{x^2}{a^2}+\frac{y^2}{b^2}$$

$Рассмотрим\ частный\ случай\ при\ a=1\ и\ b=1$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10, 10))

ax = Axes3D(fig,  auto_add_to_figure=False)
fig.add_axes(ax)

X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = meshgrid(X, Y)
Z = X**2 + Y**2

ax.plot_surface(X, Y, Z, color='pink')
```

### Двуполосный гиперболоид

$$-\frac{x^2}{a^2} -\frac{y^2}{b^2} + \frac{z^2}{c^2} = 1$$

```python
fig = plt.figure(figsize=(10, 10))
ax = Axes3D(fig,  auto_add_to_figure=False)
fig.add_axes(ax)

a = 2
b = 2
с = 11

X = np.arange(-2, 2, 0.1)
Y = np.arange(-2, 2, 0.1)

X, Y = meshgrid(X, Y)
Z = (X**2/a**2 + Y**2/b**2 + 1)**0.5 * с
ax.plot_surface(X, Y, Z, color='pink')
ax.plot_surface(X, Y, -Z, color='pink')
```
