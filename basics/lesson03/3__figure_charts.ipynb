## Окружность
$$(x-x_0)^2+(y-y_0)^2=R^2$$
$$\pm{y}=\sqrt{R^2-(x-x_0)^2}+y_0, \ |x - x_0| <=R$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def value_x(R):
    return np.linspace(-R, R, 1000)
```

```python
def my_circle(x, r, x0, y0):
    return (r**2-(x-x0)**2)**0.5+y0
```

```python
plt.figure(figsize=(10, 10))
plt.axis([-5, 5, -5, 5])

r = 4
x = value_x(r)
y = my_circle(x, r, 0, 0)
plt.plot(x, y, color='pink')
plt.plot(x, -y, color='pink')
plt.grid(color='#eeefff', linestyle='-', linewidth=1)
```

## Элипс
$$\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1, \ a\neq0, \ b\neq0$$
$$\pm{y}=\sqrt{b^2-\frac{b^2}{a^2}*x^2}$$

```python
def my_elipse(x, a, b):
    return (b**2-b**2*x**2/a**2)**0.5
```

```python
def value_x(a):
    return np.linspace(-a, a, 1000)
```

```python
plt.figure(figsize=(10, 10))
plt.axis([-9, 9, -5, 5])
a = 8
b = 4
x = value_x(a)
y = my_elipse(x, a, b)

plt.gca().set_aspect('equal')
plt.plot(x, y, color='pink')
plt.plot(x, -y, color='pink')
plt.grid(color='#eeefff', linestyle='-', linewidth=1)
```

## Гипербола

$$\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1, \ a\neq0, \ b\neq0$$
$$\pm{y}=\sqrt{\frac{b^2}{a^2}*x^2 - b^2}$$
$$y=\pm{b}*\sqrt{\frac{x^2}{a^2} - 1}$$

```python
def my_hyperbole(x, a, b):
    return b*(x**2/a**2 - 1)**0.5
```

```python
plt.figure(figsize=(10, 10))
plt.axis([-40, 40, -40, 40])

a = 5
b = 7
x = []
y = []

for i in np.arange(a, 40, 0.1):
    x.append(i)
    y.append(my_hyperbole(i, a, b))

plt.gca().set_aspect('equal')
plt.plot(x, y, color='pink')
plt.plot(x, list(map(lambda z: -z, y)), color='pink')
plt.plot(list(map(lambda z: -z, x)), list(map(lambda z: -z, y)), color='pink')
plt.plot(list(map(lambda z: -z, x)), y, color='pink')
plt.grid(color='#eeefff', linestyle='-', linewidth=1)
```
