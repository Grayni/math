### Параллельные плоскости
$$A_1*x + B_1*y + C_1*z + D_1 = 0$$
<br>
#### Плоскость проходящая через начало координат и параллельная другим с теми же коэффициентами и D = 0:
$$A_1*x + B_1*y + C_1*z = 0$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10, 10))

ax = Axes3D(fig,  auto_add_to_figure=False)
fig.add_axes(ax)

a = -1
b = .3
c = -1
d = 12
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = meshgrid(X, Y)

# Коэффициент D = 12
Z1 = (-d - a*X - b*Y) / c


a = -2
b = .6
c = -2
d = 0
# Коэффициент D = 0
Z2 = (d -a*X - b*Y) / c

ax.plot_wireframe(X, Y, Z1, color='blue')
ax.plot_wireframe(X, Y, Z2, color='red')

show()
```
