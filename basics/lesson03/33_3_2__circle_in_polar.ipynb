### Окружность в полярных координатах

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(7, 7))
def circle(R):
    T = np.linspace(-np.pi, np.pi, 360)
    plt.polar(T, np.linspace(R, R, 360), color='pink')
    plt.grid(color='#eeefff')

circle(7)
```
