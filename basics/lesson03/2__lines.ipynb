При соблюдении пропорций осей (x, y) 1:1 мы получим перпендикулярные линии при противоположных угловых коэффициентах этих линий.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5, 5))
plt.axis([-5, 5, -5, 5])

x = np.linspace(-5, 5, 21)

y2 = (-1/2) * x
plt.plot(x, y2, color='blue')


y3 = 2 * x
plt.plot(x, y3, color='green')

plt.xlabel('x')
plt.ylabel('y')
```
