### Линейная регрессия


$$Уравнение \ прямой: \\ y=ax+b$$

$$a=\frac{\Sigma x_i \Sigma y_i - n\Sigma{x_i y_i}}{(\Sigma x_i)^2-n\Sigma x_i^2}$$
$$$$
$$b=\frac{\Sigma y_i - a\Sigma x_i}{n}$$

$$Коэффициент \ корреляции:$$
$$R = \frac{\Sigma{(x_i-x_m)(y_i-y_m)}}{\sqrt{\Sigma(x_i-x_m)^2 \Sigma(y_i-y_m)^2}}$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 3, 21)
plt.plot(x, x**2 - 1, marker='o', markerfacecolor='white', markeredgecolor='#bbb', color='pink')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.2, 5)
plt.grid(color='#eee')
```

```python
n = 100
r = 0.7

x = np.random.rand(n)
y = r*x + (1-r)*np.random.rand(n)
plt.plot(x, y, 'o', markerfacecolor='white', markeredgecolor='#bbb')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color='#eee')

a = (np.sum(x)*np.sum(y) - n*np.sum(x*y))/(np.sum(x)**2 - n*np.sum(x**2))
b = (np.sum(y) - a*np.sum(x))/n
print(a, b)

A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y, rcond=None)[0]
print(a1, b1)
plt.plot([0, 1], [b, a + b])
```

```python
# Коэффициент корреляции

x_mean = np.sum(x)/n
y_mean = np.sum(y)/n

R = np.sum((x-x_mean)*(y-y_mean))/np.sqrt(np.sum((x-x_mean)**2)*np.sum((y-y_mean)**2))
print(R)
```
