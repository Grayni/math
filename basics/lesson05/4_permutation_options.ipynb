### Возможные варианты перестановок

```python
import itertools
```

```python
l = []
val = 0

# количество событий = 3, количество испытаний = 4
for p in itertools.product('012', repeat=4):
    val += 1
    l.append(p)

print(f'Количество возможных вариантов: {val}')
print(*l)

print('\n\n')

l = []
val = 0

# количество событий = 4, длина выборки = 2
for p in itertools.product('0123', repeat=2):
    val += 1
    l.append(p)

print(f'Количество возможных вариантов: {val}')
print(*l)
```
